```python
sp = lrb.sp
null_mask = ones_like(array(sp)).astype(np.bool)
pca_0 = sorters.PCATransform(sp,null_mask,defaults.pca_trans)
pca_0.transform()
km_0 = sorters.KMeansCluster(sp,null_mask,pca_0.trnsmtrx,defaults.km_cluster)
#km_0.params['init'] = km_test.est.cluster_centers_
km_0.select()
```

```python
sorters.plot_clusters(km_0,slice(0,4000,10))
display(gcf())
close()
```

```python
md_params = {'trans_dims':1}
mask = km_0.mask_from_labels('0')
pca_00 = sorters.PCATransform(sp,mask,defaults.pca_trans)
pca_00.transform()
km_00 = sorters.KMeansCluster(sp,mask,pca_00.trnsmtrx,defaults.km_cluster)
#th_00 = sorters.ThreshSelector(sp,mask,pca_00.trnsmtrx,{})
#thresh = percentile(md_00.trnsmtrx,90)
km_00.select()
```

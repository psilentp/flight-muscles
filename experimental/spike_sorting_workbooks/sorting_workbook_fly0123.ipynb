```python
import neo
import quantities as pq
import sorters
import sorting_defaults as defaults
import h5py
import flylib
import db_access as dba
```

```python
fly_db = dba.get_db()
```

```python
swarm = flylib.Squadron(fly_db)
```

```python
fly = swarm.flies[8]
print(fly.fly_num)
lrb = fly.experiments['lr_blob_expansion']
lrb.calc_spikepool()
```

```python
sp = lrb.sp
null_mask = ones_like(array(sp)).astype(np.bool)
pca_0 = sorters.PCATransform(sp,null_mask,defaults.pca_trans)
pca_0.transform()
km_0 = sorters.KMeansCluster(sp,null_mask,pca_0.trnsmtrx,defaults.km_cluster)
#km_0.params['init'] = km_test.est.cluster_centers_
km_0.select()
```

```python
sorters.plot_clusters(km_0,slice(0,4000,10))
display(gcf())
close()
```

```python
md_params = {'trans_dims':1}
mask = km_0.mask_from_labels('1')
md_01 = sorters.MedTrans(sp,mask,md_params)
md_01.transform()
th_01 = sorters.ThreshSelector(sp,mask,md_01.trnsmtrx,{})
thresh = percentile(md_01.trnsmtrx,98)
th_01.select(thresh)
```

```python
sorters.plot_clusters(th_01,slice(0,4000,10))
display(gcf())
close()
```

```python
selector = th_01
label = '0'
selection_mask = selector.mask_from_labels(label)
figure(figsize=(3,1))
l = plot(array(lrb.exp_record['spike_data']['wv_mtrx'])[selection_mask,:][:1000:10].T,color = 'k',alpha = 0.1)
display(gcf())
close()
```

```python
lrb.set_spike_selection('b1',selection_mask)
lrb.sync_sorted_spikes('b1')
```

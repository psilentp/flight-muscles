```python
import flylib
import db_access as dba
fly_db = dba.get_db()
```

```python
test_group = [217]
```

```python
swarm = flylib.Squadron(fly_db,test_group)
```

```python
fly = swarm.flies[0]
expmnt = fly.experiments.values()[1]
```

```python
stim_cond = np.array(expmnt.exp_record['tiff_data']['axon_framebase']['DAC2'])
lwing = np.array(expmnt.exp_record['tiff_data']['axon_framebase']['Ph0'])
rwing = np.array(expmnt.exp_record['tiff_data']['axon_framebase']['Ph1'])
lmr = np.array(expmnt.exp_record['tiff_data']['axon_framebase']['Ph2'])
aux = np.array(expmnt.exp_record['tiff_data']['axon_framebase']['Ph3'])
xpos = np.array(expmnt.exp_record['tiff_data']['axon_framebase']['Xpos'])
times = np.array(expmnt.exp_record['tiff_data']['axon_framebase']['times'])
```

```python
plot(aux)
flight_mask = aux>1.9
plot(flight_mask);display(gcf());close()
```

```python
yaw_90_selection_mask = around(stim_cond)==2
yaw_270_selection_mask = around(stim_cond)==3
yaw_90_trials = flylib.idx_by_thresh(yaw_90_selection_mask,0.5)
yaw_270_trials = flylib.idx_by_thresh(yaw_270_selection_mask,0.5)
plot(yaw_90_selection_mask);plot(yaw_270_selection_mask)
display(gcf());close()
```

```python
sig = rwing
trials = yaw_90_trials
trials = [x for x in trials if len(x)>0]
yaw_90_times = [times[idx]-times[idx[0]] for idx in trials if sum(flight_mask[idx])==0]
yaw_90_data = [sig[idx] for idx in trials if sum(flight_mask[idx])==0]

trials = yaw_270_trials
trials = [x for x in trials if len(x)>0]
yaw_270_times = [times[idx]-times[idx[0]] for idx in trials if sum(flight_mask[idx])==0]
yaw_270_data = [sig[idx] for idx in trials if sum(flight_mask[idx])==0]

import scipy.interpolate
fun = scipy.interpolate.griddata

yaw_270_resamp = np.array([fun(t,d,linspace(0,3,100)) for t,d in zip(yaw_270_times,yaw_270_data)])
yaw_90_resamp = np.array([fun(t,d,linspace(0,3,100)) for t,d in zip(yaw_90_times,yaw_90_data)])
plot(mean(yaw_270_resamp,axis = 0))
plot(mean(yaw_90_resamp,axis = 0))
display(gcf());close()
```

```python
def plot_asap_data(fly,bx = [115,130,21,45]):
    expmnt = fly.fly_record['experiments'].values()[0]
    imgs = expmnt['tiff_data']['images']
    wb_frequency = expmnt['tiff_data']['axon_framebase']['wb_frequency']
    tm = expmnt['tiff_data']['axon_framebase']['times']
    import pylab as plb
    plb.rcParams['pdf.fonttype'] = 42
    roi_data = imgs[:,bx[0]:bx[1],bx[2]:bx[3]]
    shape(roi_data)
    rng = slice(0,10000)
    subplot(2,1,1)
    gca().set_ylabel('roi mean')
    plot(tm[rng],np.mean(np.mean(roi_data,axis = 1),axis = 1)[rng])
    subplot(2,1,2)
    gca().set_ylabel('wb frequency')
    plot(tm[rng],wb_frequency[rng])
```

```python
def show_roi_img(fly,gain = 8,bx = [115,130,21,45]):
    expmnt = fly.fly_record['experiments'].values()[0]
    imgs = expmnt['tiff_data']['images']
    img = dstack((imgs[0],imgs[0],imgs[0]))
    img[bx[0]:bx[1],bx[2]:bx[3],0] += 10
    imshow(img*gain)
```

```python
##ASAP2s 21ms/frame
bx = [100,115,20,45]
fly = flb.NetFly(510)
fly.open_fly_record()
figure()
show_roi_img(fly,bx=bx,gain = 2)
figure()
plot_asap_data(fly,bx)
```
